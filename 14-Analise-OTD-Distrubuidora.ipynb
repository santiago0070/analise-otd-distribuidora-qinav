BLOCO 1 - Importar bibliotecas

```python
import pandas as pd
import numpy as  np
```

BLOCO 2 - Definir Quantidade de dados

```python


np.random.seed(42)
n = 2000
```

BLOCO 3 - Criar estrutura incial

```python
product_types = ['Suturas', 'Endoscópios', 'Cosméticos Médicos', 'Medicamentos', 'OPME']

df = pd.DataFrame({
    'sku':['SKU'+str(i).zfill(4) for i in range(n)],
    'tipo_produto': np.random.choice(product_types,n),
    'data':pd.date_range('2020-01-01', periods=n, freq='D')
})
```

BLOCO 4 - Criar Lead Time realista

```python
lead_map = {
    'Suturas':1.4,
    'Endoscópios':1.1,
    'Cosméticos Médicos':1.0,
    'Medicamentos':1.0,
    'OPME':0.7
}

df['prazo_entrega'] = (
    df['tipo_produto']
    .map(lead_map)
    + np.random.normal(0,1.5,n)
)


# prazo_entrega * performance
df['tempo_entrega'] = (
    df['prazo_entrega'] * df['tipo_produto'].map(lead_map) 
    + np.random.normal(0,1.5, n) # Reduzi o erro para a performance aparecer mais
)
# Arredondamentos padrão
df['prazo_entrega'] = df['prazo_entrega'].clip(1, 15).round(1)

# Arredondamentos padrão
df['tempo_entrega'] = df['tempo_entrega'].clip(1, 20).round(1)
```

BLOCO 5 — Criar estoque

```python
df['nivel_estoque'] = np.random.poisson(200,n)
```

BLOCO 6 - Vendas

```python
df['qtd_vendida'] = (
df['nivel_estoque'] * np.random.uniform(0.1,0.3,n)
).astype(int)
```

BlOCO 7 - Preço

```python
mapa_preco={
    'Suturas': 50,
    'Endoscópios': 5000,
    'Cosméticos Médicos': 150,
    'Medicamentos': 80,
    'OPME': 12000
}
df['preco_unitario'] = df['tipo_produto'].map(mapa_preco)
```

BLOCO 8 - Receita

```python
df['receita'] = df['qtd_vendida'] *  df['preco_unitario']
```

BLOCO 9 - Indicador OTD

```python
# Criando a coluna OTD_FLAG de forma clara
df['otd_flag'] = np.where(
    df['prazo_entrega'] >= df['tempo_entrega'],
    1,
    0
)
```

BlOCO 10 - Exporta CSV

```python
df.to_csv(
    'supply_chain_distribuidora_qinav.csv',
    index = False,
    sep = ',',
    encoding = 'utf-8-sig'
)
```
